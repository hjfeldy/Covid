### Get average national hospital capacity by month, and add that data to the main clean_data frame

```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/clean_data.csv')
cap = pd.read_csv('../Data/hospital_occupancy.csv', index_col = False)
```

```python
cap
```

```python
# Perform Merge
df = pd.merge(df, cap, left_on = 'month', right_index= True).rename(columns={
    'adult_icu_bed_covid_utilization': 'ICU_occupancy',
    'inpatient_beds_utilization': 'inpatient__occupancy'
})
df
```

```python
df.to_csv('../Data/clean_data.csv', index= False)
```

```python
import pandas as pd
import numpy as np
```

```python
# You can pass a url to pd.read_csv, meaning we don't have to keep a copy of the csv downloaded
df = pd.read_csv('http://data.cdc.gov/api/views/vbim-akqf/rows.csv?accessType=DOWNLOAD')
# This will take a LONG time - but it offers the advantage that ANYONE 
df
```

```python
#Because we will apply regression to this, we need represents our variables of interest as numbers

df = df.replace(['Missing', 'Unknown'], np.nan) #no interest in these 
df = df.replace('No', 0)
df = df.replace('Yes', 1)

df = df.replace('Male', 1)
df = df.replace('Female', 0)
```

```python
df = df.dropna()

df
```
